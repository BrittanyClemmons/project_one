## Analysis of Directors and their IMDB Scores

This analysis

```python

df.imdb_score.describe()
```

```python
#Create a dataframe of directors and imdb score for each movie
directors_and_imdb = pd.DataFrame(df[['director_name', 'imdb_score']])
directors_and_imdb
```

```python
#Create a DataFrame of directors and IMDB Scores in which the score is above the average

directors_and_imdb_above_avg = directors_and_imdb[directors_and_imdb['imdb_score'] > 6.5]
directors_and_imdb_above_avg.reset_index(drop=True)
```

```python
directors_and_imdb_above_avg_group = directors_and_imdb_above_avg.groupby('director_name').count()
directors_and_imdb_above_avg_group= directors_and_imdb_above_avg_group['imdb_score'].nlargest(30)
directors_and_imdb_above_avg_group
```

```python
directors_and_imdb_above_avg_group = directors_and_imdb_above_avg.groupby('director_name').mean()
directors_above_avg_count = directors_and_imdb_above_avg_group['imdb_score'].nlargest(30)
directors_count = pd.DataFrame(directors_above_avg_count)
directors_count
```

```python
directors_and_imdb_above_avg_group = directors_and_imdb_above_avg.groupby('director_name').mean()
directors_above_avg_mean = directors_and_imdb_above_avg_group['imdb_score'].nlargest(30)
directors_score = pd.DataFrame(directors_above_avg_mean)
```

```python
df_merge_director_imdb = pd.concat([directors_count, directors_score, on='director_name'])
df_merge_director_imdb
```

```python
#Create a group based on directors
directors_imdb_group = directors_and_imdb.groupby('director_name')

#Average imdb score per director
directors_avg_imdb = directors_imdb_group['imdb_score'].mean().nlargest(10)
directors_avg_imdb
```
