```python
#Import dependencies
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Import data
csv_data = pd.read_csv("Resources/movie_metadata.csv", encoding="utf-8")
df_data = pd.DataFrame(csv_data)
df_data
```

```python
#Clean dataset
df_clean = df_data.dropna().reset_index(drop=True)

df = df_clean
df.head(10)
```

```python
df.columns
```

# Exploratory Analysis

## Analysis of the Quantity of Movies Made and average IMDB Scores per Director

```python
#Create a dataframe of directors and imdb score for each movie
df_directors_imdb = pd.DataFrame(df[['director_name', 'imdb_score']])
df_directors_imdb
```

```python
#We will analyze imdb scores of above average score vs. below average score
df_directors_imdb['imdb_score'].describe()
```

The average IMDB Score is 6.47.

```python
#Group dataset by directors to find their average IMDB Score
df_directors_avg_score = df_directors_imdb.groupby(['director_name']).mean()
```

```python
df_directors_avg_score
```

```python
df_directors_avg_score.describe()
```

```python
#Group data by directors to find the count of Movies made
df_directors_movie_count = df_directors_imdb.groupby(['director_name']).count()
df_directors_movie_count
```

```python
df_directors_movie_count.describe()
```

```python
#Prep data from both dataframes to be graphed in scatter plot

#Convert avg imdb score dataframe column into a list
y = df_directors_avg_score['imdb_score'].tolist()

#Convert count imdb score dataframe column into a list
x = df_directors_movie_count['imdb_score'].tolist()
```

## Practice makes perfect? 

### As the quantity of movies made per director increases, does the director's average IMDB score increase as well?

```python
#Create a matplotlib scatter plot based on the directors data above

#Set upper and lower limits
plt.xlim(min(x)-1,max(x)+ 0.5)
plt.ylim(min(y)-1, max(y)+0.5)

#Set title and chart labels
plt.title("Mean of IMDB Score over Number of Movies Made per Director")
plt.xlabel("Number of Movies made per Director")
plt.ylabel("Mean IMDB Score per Director")

#Scatter plot data
plt.scatter(x, y, marker='o', facecolors='red', edgecolors='black')

#Trendline
z = np.polyfit(x,y,1)
p = np.poly1d(z)
plt.plot(x, p(x), "b--")

#Trendline equation
print("y=%.6fx+(%.6f)"%(z[0],z[1]))

#Save Scatter Plot
plt.savefig("Images/Mean IMDB over Number of Movies per Director")
plt.show()
```

The scatter plot above shows a weak positive correlation between the quantity of movies made and the average IMDB score per Director

## Does hiring a top actor guarantee movie success?

## Analysis of Top 10 Popular Actors and average IMDB Scores

```python
#Create a dataframe of all movies, actors, movie facebook likes

df_top_actors = pd.DataFrame(df[["actor_1_name", 'movie_title']])
df_top_actors
```

```python
#Group movies by actors and sum the total facebook likes for all of the actor's movies
#This finds the top 10 popular actors by count of appearance as lead actor in movies

top_actors_group = df_top_actors.groupby('actor_1_name',as_index=False).count().nlargest(10, 'movie_title').reset_index(drop=True)
top_actors_group
```

```python
#Find all movies with only top 10 actors found above

movies_top_10_actors = df[(df['actor_1_name'] == top_actors_group.iloc[0,0]) |
                          (df['actor_1_name'] == top_actors_group.iloc[1,0]) |
                          (df['actor_1_name'] == top_actors_group.iloc[2,0]) |
                          (df['actor_1_name'] == top_actors_group.iloc[3,0]) |
                          (df['actor_1_name'] == top_actors_group.iloc[4,0]) |
                          (df['actor_1_name'] == top_actors_group.iloc[5,0]) |
                          (df['actor_1_name'] == top_actors_group.iloc[6,0]) |
                          (df['actor_1_name'] == top_actors_group.iloc[7,0]) |
                          (df['actor_1_name'] == top_actors_group.iloc[8,0]) |
                          (df['actor_1_name'] == top_actors_group.iloc[9,0])
                         ]
movies_top_10_actors['actor_1_name'].unique()
```

```python
#Create dataFrame of movies with top 10 actors and IMDB scores

df_actors_imdb = pd.DataFrame(movies_top_10_actors[['actor_1_name', 'imdb_score']]).reset_index(drop=True)
df_actors_imdb
```

```python
#Group dataset by actor 1 to find their average IMDB Score
#Alphabetical order
df_actors_avg_imdb = df_actors_imdb.groupby(['actor_1_name'], as_index=False).mean()
df_actors_avg_imdb
```

```python
#Bring back actors movie count in alphabetical order
df_top_actors_count = pd.DataFrame(top_actors_group).sort_values(by=['actor_1_name']).reset_index(drop=True)
df_top_actors_count
```

```python
#Prep data from both dataframes to be graphed in scatter plot

#Convert avg imdb score dataframe column into a list
b = df_actors_avg_imdb['imdb_score'].tolist()

#Convert count imdb score dataframe column into a list
a = df_top_actors_count['movie_title'].tolist()
```

```python
#Create a matplotlib scatter plot based on the directors data above

#Set upper and lower limits
plt.xlim(min(a)-0.25,max(a)+ 0.25)
plt.ylim(min(b)-0.25, max(b)+0.25)

#Set title and chart labels
plt.title("Mean of IMDB Score over Number of Movie Appearance by Top 10 Actors")
plt.xlabel("Number of Movies by Top 10 Actors")
plt.ylabel("Mean IMDB Score")

#Scatter plot data
plt.scatter(a, b, marker='o', facecolors='red', edgecolors='black')

#Trendline
z = np.polyfit(a,b,1)
p = np.poly1d(z)
plt.plot(a, p(a), "b--")

#Trendline equation
print("y=%.6fx+(%.6f)"%(z[0],z[1]))

#Save Scatter Plot
plt.savefig("Images/Mean of IMDB Score over Number of Movie Appearance by Top 10 Actors")
plt.show()
```
