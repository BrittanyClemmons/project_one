```python
#Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


#Import data
csv_data = pd.read_csv("Resources/movie_metadata.csv", encoding="utf-8")
df = pd.DataFrame(csv_data)
df
```

```python
df.count()
```

```python
df_clean=df.dropna()
```

```python
df_clean
```

```python
df_clean.columns
```

```python
#Duration vs IMBD Scores
e = df_clean[["movie_title", "duration", "imdb_score"]]
e
```

```python
x = e['duration']
y = e['imdb_score']

plt.scatter(x=e['duration'], y=e['imdb_score'])
plt.title("Duration impact on IMDB Score")
plt.xlabel("duration (minutes)")
plt.ylabel("imdb_score")
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
plt
```

```python
f = df_clean["aspect_ratio"].value_counts()
f
```

```python
f = df_clean[["aspect_ratio", "budget", "gross"]]
f['profit'] = f["gross"] - f["budget"]

aspect_1 = f.loc[(f["aspect_ratio"] == 2.35)]
aspect_2 = f.loc[(f["aspect_ratio"] == 1.85)]
aspect_3 = f.loc[(f["aspect_ratio"] == 1.37)]
aspect_4 = f.loc[(f["aspect_ratio"] == 1.66)]
aspect_5 = f.loc[(f["aspect_ratio"] == 1.78)]

aspect_1_mean = aspect_1["profit"].mean()*10
aspect_2_mean = aspect_2["profit"].mean()*10
aspect_3_mean = aspect_3["profit"].mean()*10
aspect_4_mean = aspect_4["profit"].mean()*10
aspect_5_mean = aspect_5["profit"].mean()*10

aspect_1_sem = aspect_1["profit"].sem()
aspect_2_sem = aspect_2["profit"].sem()
aspect_3_sem = aspect_3["profit"].sem()
aspect_4_sem = aspect_4["profit"].sem()
aspect_5_sem = aspect_5["profit"].sem()

aspect_list_mean = [aspect_1_mean, aspect_2_mean, aspect_3_mean, aspect_4_mean, aspect_5_mean]
aspect_list_sem = [aspect_1_sem, aspect_2_sem, aspect_3_sem, aspect_4_sem, aspect_5_sem]

x_axis = np.arange(len(aspect_list_mean))
plt.bar(x_axis, aspect_list_mean,  yerr = aspect_list_sem, color='r', alpha=0.5, align="center", capsize=10)
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["2.35", "1.85", "1.37", "1.66", "1.78"])


plt.title("Aspect Ratio Impact on Profit ")
plt.xlabel("Aspect Ratio")
plt.ylabel("Mean Profit (in hundreds of millions)")

plt
```
