```python
#Calculate profit and add column to existing data frame
df_clean['profit'] = df_clean["gross"] - df_clean["budget"]
profit_df = df_clean
a = profit_df[["movie_title", "movie_facebook_likes", "profit"]]
b = a.sort_values("profit", ascending=False)
c = b.nlargest(50, "profit")
c
```

```python
plt.scatter(x=c['movie_facebook_likes'], y=c['profit'])
plt.title("Movie Popularity vs Profit")
plt.xlabel("Facebook Likes")
plt.ylabel("Profit (in hundreds of millions)")
plt
```
